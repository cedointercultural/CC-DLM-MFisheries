```python
import numpy as np
import pandas as pd
import shap
import tensorflow as tf
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Cargar el modelo MoE entrenado
model_path = "modelos_moe/JAIBA_cluster_7_moe_model.h5"
model = load_model(model_path)

# Cargar los datos utilizados para entrenar
train_data_path = "data/data.csv"  # Cambiar al path real si es diferente

# Cargar los datos de entrenamiento
df_train = pd.read_csv(train_data_path)

# Seleccionar las características de entrada basadas en los nombres de columnas utilizados previamente
features_train = ['landed_w_kg', 'Cluster_Label', 'mean_temp_30m', 'mean_temp_10m', 'thetao_sfc=6',
                  'thetao_sfc=7.92956018447876', 'thetao_sfc=9.572997093200684', 'thetao_sfc=11.40499973297119',
                  'thetao_sfc=13.46714019775391', 'thetao_sfc=15.8100700378418', 'thetao_sfc=18.49555969238281',
                  'thetao_sfc=21.59881973266602', 'thetao_sfc=25.21141052246094', 'thetao_sfc=29.44473075866699']

# Seleccionar los datos de características de entrenamiento
X_train = df_train[features_train].values

# Ajustar el reshape con una secuencia temporal acorde con el tamaño del dataset
num_timesteps = 6  # Utilizaremos 6 pasos de tiempo
n_features = len(features_train)  # Número de características

# Tomar los primeros 6 registros del conjunto de datos de entrenamiento para la primera predicción
initial_input = X_train[:num_timesteps].reshape(1, num_timesteps, n_features)

# Almacenar las predicciones futuras
future_predictions = []
shap_values_list = []

# Crear una función envolvente para que el modelo pueda recibir entradas de 2D y transformarlas en 3D
def predict_wrapper(input_data):
    input_data_reshaped = input_data.reshape(-1, num_timesteps, n_features)
    return model.predict(input_data_reshaped).flatten()  # Asegurarse de devolver una salida plana

# Inicializar KernelExplainer con una muestra del conjunto de datos
background = X_train[-(num_timesteps * 10):]  # Tomar los últimos 10 bloques de tamaño `num_timesteps`
background = background.reshape(-1, num_timesteps * n_features)  # Aplanar las muestras de fondo para KernelExplainer

explainer = shap.KernelExplainer(predict_wrapper, background)

# Pronóstico secuencial utilizando solo los datos de entrenamiento
num_predictions = 10  # Número de predicciones secuenciales que queremos realizar
for i in range(num_predictions):
    # Predecir el próximo valor usando el modelo
    next_prediction = model.predict(initial_input)

    # Asegurarse de que la predicción tenga la forma adecuada
    next_prediction_value = next_prediction.flatten()[0]  # Extraer el valor predicho del desembarque (landed_w_kg)

    # Crear la nueva entrada para el siguiente paso de tiempo:
    # Desplazar los valores hacia atrás en `initial_input` y agregar `next_prediction_value`
    next_input = np.concatenate([initial_input[:, 1:, :], np.array([[[next_prediction_value] + list(initial_input[0, -1, 1:])]])], axis=1)
    initial_input = next_input

    # Almacenar la predicción
    future_predictions.append(next_prediction_value)

    # Calcular los valores SHAP para la predicción actual
    input_flat = initial_input.reshape(1, -1)  # Aplanar la entrada para que sea compatible con KernelExplainer
    shap_values = explainer.shap_values(input_flat, nsamples=50)  # Reducir el número de muestras para mejorar el rendimiento
    shap_values_list.append(shap_values)

# Convertir las predicciones futuras a un array de numpy
future_predictions = np.array(future_predictions)

# Mostrar los gráficos de valores SHAP para la primera predicción
shap.summary_plot(shap_values_list[0], input_flat, plot_type="bar")
```

```python
# Listar los nombres reales de las características
feature_names = ['landed_w_kg', 'Cluster_Label', 'mean_temp_30m', 'mean_temp_10m', 'thetao_sfc=6',
                 'thetao_sfc=7.92956018447876', 'thetao_sfc=9.572997093200684', 'thetao_sfc=11.40499973297119',
                 'thetao_sfc=13.46714019775391', 'thetao_sfc=15.8100700378418', 'thetao_sfc=18.49555969238281',
                 'thetao_sfc=21.59881973266602', 'thetao_sfc=25.21141052246094', 'thetao_sfc=29.44473075866699']

# Crear una lista de nombres para todas las características en la entrada aplanada
# Dado que estamos utilizando una entrada aplanada, los nombres se repetirán para cada paso temporal
flattened_feature_names = []
for timestep in range(num_timesteps):
    for feature in feature_names:
        flattened_feature_names.append(f"{feature}_timestep_{timestep+1}")

# Mostrar la gráfica SHAP con los nombres correctos
shap.summary_plot(shap_values_list[0], feature_names=flattened_feature_names, plot_type="bar")
```

```python
import numpy as np
import pandas as pd
import joblib
import tensorflow as tf
import shap
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Filtrar los datos por especie y clúster
species_to_filter = "JAIBA"  # Especie de interés
cluster_to_filter = 7            # Clúster de interés

# Cargar los datos históricos y futuros
data = pd.read_csv('data/data.csv')
future_data = pd.read_csv('future_data.csv')

# Filtrar los datos según la especie y el clúster
filtered_data = data[(data['species'] == species_to_filter) & (data['Cluster_Label'] == cluster_to_filter)]

# Cargar el modelo y el scaler
model_path = 'modelos_moe/JAIBA_cluster_7_moe_model.h5'  # Cambiar al path real si es diferente
scaler_path = 'modelos_moe/JAIBA_cluster_7_moe_scaler.pkl'  # Cambiar al path real si es diferente
model = load_model(model_path)
scaler = joblib.load(scaler_path)

# Seleccionar las características de entrada
features = ['landed_w_kg', 'Cluster_Label', 'mean_temp_30m', 'mean_temp_10m', 'thetao_sfc=6',
            'thetao_sfc=7.92956018447876', 'thetao_sfc=9.572997093200684', 'thetao_sfc=11.40499973297119',
            'thetao_sfc=13.46714019775391', 'thetao_sfc=15.8100700378418', 'thetao_sfc=18.49555969238281',
            'thetao_sfc=21.59881973266602', 'thetao_sfc=25.21141052246094', 'thetao_sfc=29.44473075866699']

# Escalar los datos de entrenamiento
filtered_data_scaled = scaler.transform(filtered_data[features])

# Preparar los datos para el pronóstico
num_timesteps = 6
n_features = len(features)
X_input = filtered_data_scaled[-num_timesteps:].reshape(1, num_timesteps, n_features)

# Pronosticar usando el modelo cargado
future_predictions = []
for i in range(len(future_data)):
    # Realizar la predicción
    next_prediction = model.predict(X_input)
    future_predictions.append(next_prediction[0][0])
    
    # Actualizar la entrada para la siguiente predicción
    future_temp_values = future_data.iloc[i][['depth_mean_30m', 'depth_mean_10m', 'depth_6.00', 'depth_7.93',
                                              'depth_9.57', 'depth_11.40', 'depth_13.47', 'depth_15.81',
                                              'depth_18.50', 'depth_21.60', 'depth_25.21', 'depth_29.44']].values
    future_temp_values = future_temp_values.reshape(1, -1)  # Asegurar que los datos futuros tengan la misma dimensión
    new_record = np.concatenate((next_prediction.reshape(1, 1), np.array([[cluster_to_filter]]), future_temp_values), axis=1)
    new_record_scaled = scaler.transform(new_record)
    X_input = np.concatenate((X_input[:, 1:, :], new_record_scaled.reshape(1, 1, -1)), axis=1)

# Análisis SHAP
# Crear una función envolvente para el modelo
explainer_input = filtered_data_scaled[-(num_timesteps * 10):].reshape(-1, num_timesteps * n_features)

def predict_wrapper(input_data):
    input_data_reshaped = input_data.reshape(-1, num_timesteps, n_features)
    return model.predict(input_data_reshaped).flatten()

# Inicializar KernelExplainer
explainer = shap.KernelExplainer(predict_wrapper, explainer_input)

# Calcular los valores SHAP para una muestra del conjunto de entrenamiento
shap_values = explainer.shap_values(explainer_input[:50], nsamples=50)

# Crear una lista de nombres para todas las características en la entrada aplanada
flattened_feature_names = []
for timestep in range(num_timesteps):
    for feature in features:
        flattened_feature_names.append(f"{feature}_timestep_{timestep + 1}")

# Mostrar la gráfica SHAP con los nombres correctos
shap.summary_plot(shap_values, feature_names=flattened_feature_names, plot_type="bar")

# Mostrar las predicciones futuras
plt.figure(figsize=(10, 6))
plt.plot(future_predictions, label='Predicted Landed Weight (kg)', marker='o')
plt.xlabel('Time Step')
plt.ylabel('Landed Weight (kg)')
plt.title(f'Predictions for {species_to_filter} in Cluster {cluster_to_filter}')
plt.legend()
plt.grid(True)
plt.show()
```
